### 3. Pre-Process Data

#### Fill NA values with the mean value of the column

```python
filled = pp.na_fill(data, fill_method = np.mean)
filled.head()
```

### 4. Generate features

#### Discretize the MonthlyIncome, DebtRatio, and NumberOfOpenCreditLinesAndLoans columns

```python
binned_data = pp.feat_mult_disc(filled, {'MonthlyIncome': [0, .25, .5, .75, 1.], 'NumberOfOpenCreditLinesAndLoans':[0, .25, .5, .75, 1.], 'DebtRatio': [0, .25, .5, .75, 1.]}, qt=True)
```

#### Turn Zipcode and other discretized variables into binary variables (one-hot encoded into columns)

```python
final_data = pp.feat_binary(binned_data, ['zipcode', 'MonthlyIncome_binned', 'NumberOfOpenCreditLinesAndLoans_binned', 'DebtRatio_binned'])
```

```python
final_data.columns
```

### 5. Build Classifier

#### I am building a decision tree classifier using most of the columns as features and the 'SeriousDlqin2yrs' as the label.

```python
feature_col = ['RevolvingUtilizationOfUnsecuredLines',
       'age', 'NumberOfTime30-59DaysPastDueNotWorse',
       'NumberOfTimes90DaysLate', 'NumberRealEstateLoansOrLines',
       'NumberOfTime60-89DaysPastDueNotWorse', 'NumberOfDependents',
       'zipcode_60601', 'zipcode_60618', 'zipcode_60625', 'zipcode_60629',
       'zipcode_60637', 'zipcode_60644', 'MonthlyIncome_binned_(-0.001, 3800.0]',
       'MonthlyIncome_binned_(3800.0, 6418.0]',
       'MonthlyIncome_binned_(6418.0, 7245.5]',
       'MonthlyIncome_binned_(7245.5, 1794060.0]',
       'NumberOfOpenCreditLinesAndLoans_binned_(-0.001, 5.0]',
       'NumberOfOpenCreditLinesAndLoans_binned_(5.0, 8.0]',
       'NumberOfOpenCreditLinesAndLoans_binned_(8.0, 11.0]',
       'NumberOfOpenCreditLinesAndLoans_binned_(11.0, 56.0]',
       'DebtRatio_binned_(-0.001, 0.176]', 'DebtRatio_binned_(0.176, 0.37]',
       'DebtRatio_binned_(0.37, 0.866]', 'DebtRatio_binned_(0.866, 106885.0]']
label_col = 'SeriousDlqin2yrs'
```

#### Split the data into testing and training sets.

```python
x_train, x_test, y_train, y_test = pp.split_data(final_data, feature_col, label_col, seed = seed) 
```

#### Build the tree with some default inputs: 

criteria = 'entropy'

depth = 5

min_leaf = 500.

```python
tree = pp.build_dtree(x_train, y_train, seed = seed)
```

#### Create predictions of probability of data point being 1 using the test data.

```python
pred = pp.predictpr_dtree(tree, x_test)
```

```python
pred
```

### 6. Evaluate (measured by accuracy using default threshold = 0.5) and visualize

```python
pp.accuracy_pred(y_test, pred)
```

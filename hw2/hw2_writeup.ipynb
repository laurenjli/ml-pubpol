### 4. Generate features

#### Discretize the MonthlyIncome, DebtRatio, and NumberOfOpenCreditLinesAndLoans columns

```python
binned_data = pp.feat_mult_disc(filled, {'MonthlyIncome': [0, .25, .5, .75, 1.], 'NumberOfOpenCreditLinesAndLoans':[0, .25, .5, .75, 1.], 'DebtRatio': [0, .25, .5, .75, 1.]}, qt=True)
```

```python
binned_data.DebtRatio.iloc[0]
```

#### Turn Zipcode and other discretized variables into a binary variable encoded into two columns

```python
final_data = pp.feat_binary(binned_data, ['zipcode', 'MonthlyIncome', 'NumberOfOpenCreditLinesAndLoans', 'DebtRatio'])
```

```python
final_data.columns
```

### 5. Build Classifier: Decision Tree

```python
feature_col = ['RevolvingUtilizationOfUnsecuredLines',
       'age', 'NumberOfTime30-59DaysPastDueNotWorse',
       'NumberOfTimes90DaysLate', 'NumberRealEstateLoansOrLines',
       'NumberOfTime60-89DaysPastDueNotWorse', 'NumberOfDependents',
       'zipcode_60601', 'zipcode_60618', 'zipcode_60625', 'zipcode_60629',
       'zipcode_60637', 'zipcode_60644', 'MonthlyIncome_(-0.001, 3800.0]',
       'MonthlyIncome_(3800.0, 6418.0]', 'MonthlyIncome_(6418.0, 7245.5]',
       'MonthlyIncome_(7245.5, 1794060.0]',
       'NumberOfOpenCreditLinesAndLoans_(-0.001, 5.0]',
       'NumberOfOpenCreditLinesAndLoans_(5.0, 8.0]',
       'NumberOfOpenCreditLinesAndLoans_(8.0, 11.0]',
       'NumberOfOpenCreditLinesAndLoans_(11.0, 56.0]',
       'DebtRatio_(-0.001, 0.176]', 'DebtRatio_(0.176, 0.37]',
       'DebtRatio_(0.37, 0.866]', 'DebtRatio_(0.866, 106885.0]']
label_col = 'SeriousDlqin2yrs'
```

```python
x_train, x_test, y_train, y_test = pp.split_data(final_data, feature_col, label_col) 
```

```python
tree = pp.build_dtree(x_train, y_train)
```

```python
pred = pp.predict_dtree(tree, x_test)
```

### 6. Evaluate (measured by accuracy) and visualize

```python
pp.accuracy(y_test, pred)
```
